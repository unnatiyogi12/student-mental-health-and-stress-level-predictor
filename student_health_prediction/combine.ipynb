```python
import joblib

student_model = joblib.load("student_stress_model.pkl")
health_model = joblib.load("health_stress_model.pkl")
```

```python
import pandas as pd

import pandas as pd

student_input = pd.DataFrame({
    "Physical Activity (hrs/week)": [4],
    "Screen Time (hrs/day)": [10],
    "Sleep Duration (hrs)": [4],
    "Age": [21],
    "Education Level": [15]
})

prediction = student_model.predict(student_input)

print("Student Stress Prediction:", prediction[0])
```

```python
test_input = pd.DataFrame({
    "Gender":[0],          # encoded value
    "Age":[20],
    "Occupation":[1],      # encoded value
    "Sleep Duration":[8],
    "Quality of Sleep":[10],
    "Physical Activity Level":[20],
    "Heart Rate":[85],
    "Daily Steps":[3500],
    "Systolic_BP":[100],
    "Diastolic_BP":[95],
})
prediction2 = health_model.predict(test_input)

print("Health Stress Prediction:", prediction2[0])
```

```python
final_stress = max(prediction[0], prediction2[0])

print("Final Stress Level:", final_stress)
```

```python
def recommend(stress):

    if stress == "High":
        return "Reduce screen time, increase physical activity, sleep 7–8 hours."
    
    elif stress == "Medium":
        return "Maintain balanced lifestyle and monitor stress regularly."
    
    else:
        return "Lifestyle looks healthy. Keep it up!"

print(recommend(final_stress))
```

```python
def detect_causes(data):

    causes = []

    if data["Screen Time (hrs/day)"] > 6:
        causes.append("High Screen Time")

    if data["Sleep Duration (hrs)"] < 6:
        causes.append("Poor Sleep")

    if data["Physical Activity (hrs/week)"] < 2:
        causes.append("Low Physical Activity")

    if data["Daily Steps"] < 4000:
        causes.append("Low Daily Steps")

    return causes
```

```python
def recommendations(data):

    tips = []

    if data["Screen Time (hrs/day)"] > 6:
        tips.append("Reduce screen time below 5 hours")

    if data["Sleep Duration (hrs)"] < 7:
        tips.append("Sleep at least 7–8 hours")

    if data["Physical Activity (hrs/week)"] < 3:
        tips.append("Exercise 30 minutes daily")

    if data["Daily Steps"] < 5000:
        tips.append("Try to walk 6000–8000 steps daily")

    return tips
```
